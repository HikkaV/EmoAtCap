```python
for i in tqdm.tqdm(pathes):
    parts_path = i.split('/')
    imname = parts_path[-1]
    if imname.startswith('.'):
        new_imname = imname[2:]
        parts_path = parts_path[:-1]
        parts_path.append(new_imname)
        shutil.move(i, '/'.join(parts_path))
```

# resizing images

```python
import tqdm
```

```python
size = 300, 400
errors=0
for i in tqdm.tqdm(pathes):
    try:
        im = Image.open(i)
    except:
        errors+=1
    try:
        im.thumbnail(size, Image.ANTIALIAS)
        im.save(i, "JPEG", quality=95)
    except:
        pass
```

```python
print(errors/len(pathes))
```

```python
for path in pathes:
    tmp_path = path.split('/')
    name = tmp_path[-1]
    shutil.move(path, os.path.join('../../data/imdb/imdb/',name))
```
